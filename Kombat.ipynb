```python
import random, math
class Personnage:
    
    def __init__(self, nom:str):
        self.nom = nom
        self.point_de_vie = 20
        self.__determiner_force()

    
    def diminuer_point_de_vie(self, impact: int, adversaire:Personnage):
        print(".................")
        if impact < 0:
            impact = impact*-1
            print(f"DEFENSE SURPUISSANTE DE {adversaire.nom}")
            print(f"{self.nom}: -{impact} point(s)")
        elif impact > 0:
            print(f"{adversaire.nom} ATTEINT: -{impact} points")
        else:
            print(f"ATTAQUE IMPUISSANT AUCUN IMPACT")
        self.point_de_vie -= impact
        

    
    def __determiner_force(self):
        self.force = math.floor(random.random()*11)
        if self.force < 5:
            self.force = 5
        

    def etat(self):
        print(f"{self.nom}: {self.point_de_vie} vie(s)")

    def attaquer(self, adversaire:Personnage):
        force_dattaque = math.floor(random.random()*self.force)

        if force_dattaque == 0:
            force_dattaque = 1
        
        print(f"Attaque de {self.nom} sur {adversaire.nom} avec une force d'attaque de {force_dattaque}")
        force_de_defense_adverse = adversaire.__se_defendre()

        impact_attaque = force_dattaque - force_de_defense_adverse

        self.diminuer_point_de_vie(impact_attaque, adversaire)

        if random.random() > 0.5:
            self.__boire_potion()

        
        
    def __se_defendre(self):
        force_de_defense = math.floor(random.random()*self.force)
        print(f"Defense de {self.nom} avec une force resitante de {force_de_defense}")
        return force_de_defense
        

    def __boire_potion(self):
        if self.point_de_vie <= 10:
            self.point_de_vie +=1
            print(f"{self.nom}: Potion magique +1 vie")
        

class Combat:
    def __init__(self, joueur1:Personnage, joueur2:Personnage):
        self.joueur1 = joueur1
        self.joueur2 = joueur2
        self.duree = 60 #60 secondes

    
    def presentation(joueur:Personnage):
        return f"{joueur.nom.capitalize()},Force de frappe maximum: {joueur.force}"
    
    def start(self):
        joueur1 = self.joueur1
        joueur2 = self.joueur2
        print(f"{joueur1.nom.capitalize()} vs {joueur2.nom.capitalize()}")
        # self.presentation(joueur1)
        # self.presentation(joueur2)
        print("READYY ??? GO")

        while self.duree > 0:
            joueur1.attaquer(joueur2)
            joueur2.attaquer(joueur1)
            print(f"{joueur1.nom}: {joueur1.point_de_vie} points")
            print(f"{joueur2.nom}: {joueur2.point_de_vie} points")
            self.duree -=1

            if joueur1.point_de_vie <=0 and joueur2.point_de_vie > 0:
                print("GAME OVER")
                print(f"VICTOIRE DE {joueur2.nom}")
                break
            
            if joueur2.point_de_vie <=0 and joueur1.point_de_vie > 0:
                print("GAME OVER")
                print(f"VICTOIRE DE {joueur1.nom}")
                break
            
        

john = Personnage("John")
doe = Personnage("Doe")
combat = Combat(john, doe)
combat.start()
```
