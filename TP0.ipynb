```python
import random
class Personnage:
    def __init__(self,nom, point_de_vie, chance):
        self.nom = nom
        self.vie = point_de_vie
        self.chance = chance
        self.__limiter_chance()
    
    def get_etat(self):
        return self.vie
    
    def get_nom(self):
        return self.nom
    
    def perd_vie(self):
        random_number = random.random()
        nb_point = 0
        if random_number * 10 > self.chance:
            if random.random() < 0.5:
                nb_point = 1
            else:
                nb_point = 2
        else:
            print(f"{self.get_nom()} a de la chance il ne perd pas de vie")
        
        self.vie -= nb_point
        if nb_point != 0:
            print(f"{self.get_nom()} perd {nb_point} points")

    def __limiter_chance(self):
        if self.chance > 4:
            self.chance = 4
        if self.chance < 1:
            self.chance = 1

    def boire_potion(self):
        self.vie +=1


def game(joueur1:Personnage, joueur2:Personnage):
    print(joueur1.chance, joueur2.chance)
    
    while joueur1.get_etat() > 0 and joueur2.get_etat() > 0:
        joueur1.perd_vie()
        joueur2.perd_vie()

    if joueur1.get_etat() <= 0 and joueur2.get_etat() > 0:
        msg = f'''{joueur2.get_nom()} est vainqueur il lui reste {joueur2.get_etat()} points, {joueur1.get_nom()} est mort'''

    elif joueur2.get_etat() <= 0 and joueur1.get_etat() >0:
        msg = f'''{joueur1.get_nom()} est vainqueur, il lui reste encore {joueur1.get_etat()} points, {joueur2.get_nom()} est mort'''

    else:
        msg = f'''Aucun vainqueur'''
        
    return msg

arthur = Personnage("Arthur", 20,5)
haaland = Personnage("Haaland", 20,-1)
game(arthur, haaland)
```
