```python
import random
class Personnage:
    def __init__(self, point_de_vie):
        self.vie = point_de_vie
    
    def get_etat(self):
        return self.vie
    
    def perd_vie(self):
        if random.random() < 0.5:
            nb_point = 1
        else:
            nb_point = 2
        self.vie -= nb_point

    def boire_potion(self):
        self.vie +=1


def game():
    bilbo = Personnage(20)
    gollum = Personnage(20)

    while bilbo.get_etat() > 0 and gollum.get_etat() > 0:
        bilbo.perd_vie()
        gollum.perd_vie()

    if bilbo.get_etat() <= 0 and gollum.get_etat() > 0:
        msg = f'''Gollum est vainqueur il lui reste {gollum.get_etat()} points, Bilbo est mort'''

    elif gollum.get_etat() <= 0 and bilbo.get_etat() >0:
        msg = f'''Bilbo est vainqueur, il lui reste encore {bilbo.get_etat()} points, que Gollum est mort'''

    else:
        msg = f'''Aucun vainqueur'''
        
    return msg

game()
```
