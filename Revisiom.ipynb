## Les listes


```python
pair = [2,4,6,8,10]
impair = [1,3,5,7,9,11]

## operations sur les listes
print(len(pair)) ## la longueur de la liste
print(pair[-1]) ## acceder aux derniers elements de la liste avec un chiffre negatif

nombres = pair + impair ## concatenation
print(nombres)
deuxfois = 2* pair ## concatenation repetee
print(deuxfois)

users = [
    {
        "id": 1,
        "username": "galandnoah",
        "email": "galandnoah@gmail.com",
        "password": "qwerty"
    },
      {
        "id": 2,
        "username": "salomondeponpays",
        "email": "deponpaays@gmail.com",
        "password": "kaloudebordo"
    },
]

## ajouter un nouvel utilisateur
users.append({
    "id": 3,
    "username": "asseck",
    "email": "asseck@gmail.com",
    "password": "numberone"
})

print(users)

print(len(users))

computer_brand =  ["apple", "lenovo", "dell", "hp", "asus", "microsoft surface", "acer", "chrome book"]
computer_brand.insert(0, "Tecno") ## inserer un element a l'indice passee en parametres
print(computer_brand)

try:
    is_apple_include = computer_brand.index("applee") ## position de la premiere occurence de "applee"
    print(is_apple_include)
except Exception as e:
    print(e)

sentence = "I ate banana yesterday"

occurence_of_a = sentence.count("a") ## nombre d'occurence de a dans la phrase
print(f" dans la phrase \" {sentence} \" \"a\" se repete {occurence_of_a} fois")





for u in users:
    if u.get("id") == 1:
        users.remove(u) ## retirer un element de la liste

nombres.sort() ## trier par ordre croissant par defaut
deuxfois.sort() 

del(computer_brand[4]) ## suppression de l'attribut d'index 4

deuxfois.sort(reverse=True) ## trie par ordre decroissant
print(deuxfois) 

print(computer_brand)

print(nombres)
print(deuxfois)

print(users)
suppended_brand = computer_brand.pop(2) ## supprimer et renvoie l'element a l'index passe en parametres
print(suppended_brand)
```

## Liste suite

```python
letters = ["a", "b", "c", "d", "e"]
copy = letters.copy() ## copie d'une liste avec reference c'est a dire que la modification d'un des deux listes avec l'autre
print(copy)
print(letters)

copy.remove("b")
print(copy)
print(letters)

copy.append("f")
print(copy)
print(letters)
```
