# Exercice 3: Vecteur 2D: surchage

```python
import math

class Vecteur2D:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def __str__(self):
       return f"""Vecteur d'abcisse {self.x} et d'ordonnee {self.y}"""

    def distance(self, vecteur):
        distance = math.sqrt(math.pow((vecteur.x - self.x),2) + math.pow((vecteur.y - self.y),2) )
        return distance

v1 = Vecteur2D(12, 10)
v2 = Vecteur2D(5, 6)
print(v1.distance(v2))
```

# Exercice 4: Heritage simple

```python
class Rectangle:
    def __init__(self, longueur, largeur, nom = "Rectangle"):
        self.longueur = longueur
        self.largeur = largeur
        self.nom = nom
    def surface(self):
        return self.largeur * self.longueur
    
class Carre(Rectangle):
    def __init__(self, cote, nom="Carre"):
        super().__init__(cote, cote, nom)

rectangle = Rectangle(20, 12)
print(rectangle.surface())
carre = Carre(10)
print(carre.surface())
```

# Exercice 5: Points

```python
class Point:

    def __init__(self, x = 0.0, y = 0.0):
        self.x = x
        self.y = y
    
    def __str__(self):
        return f"({self.x},{self.y})"

class Segment:

    def __init__(self, x1,x2, y1, y2):
        self.orig = Point(x1, y1)
        self.extrem = Point(x2, y2)

    def __str__(self):
        return f"Segment forme par les points {self.orig} et {self.extrem}"
    
segment = Segment(1,2,3,4)
print(segment)
```
